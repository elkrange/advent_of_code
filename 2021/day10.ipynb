```python
token_open = {
    '(': ')',
    '[': ']',
    '{': '}',
    '<': '>',
}
token_close = {
    ')': '(',
    ']': '[',
    '}': '{',
    '>': '<',
}
token_value = {
    ')': 3,
    ']': 57,
    '}': 1197,
    '>': 25137,
}
missing_token_value = {
    ')': 1,
    ']': 2,
    '}': 3,
    '>': 4,    
}
```

```python
lines = actual_input.split()
code = [list(line) for line in lines]
part1_score = 0
part2_lines_scores = []
for line_no, line in enumerate(code):
    sym_queue = deque()
    for pos, symbol in enumerate(line):
        if symbol in token_open.keys():
            sym_queue.append(symbol)
        elif symbol in token_close.keys():
            if token_close[symbol] == sym_queue[-1]:
                sym_queue.pop()
            else:
                part1_score += token_value[symbol]
                sym_queue = None
                break
    if sym_queue:
        closing_symbols = [token_open[symbol] for symbol in sym_queue]
        closing_symbols.reverse()
        line_score = 0
        for missing in closing_symbols:
            line_score *= 5
            line_score += missing_token_value[missing]
        part2_lines_scores.append(line_score)
num_line_scores = len(part2_lines_scores)
mid_point = num_line_scores // 2
part2_lines_scores.sort()
print(f"Part 1 score is {part1_score}")
print(f"Part 2 score is {part2_lines_scores[mid_point]}")
```
