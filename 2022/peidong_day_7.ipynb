```python
from collections import defaultdict
moves = input_data

path = [] # chained items represent the dirctory chain
memo = defaultdict(int)

for i in moves:
    seg = i.split()
    
    if seg[1] == 'cd':
        if seg[2] != '..':
            path.append(seg[2])
        else:
            path.pop()
    elif seg[1] == 'ls' or seg[0] == 'dir':
        continue
    elif seg[0].isnumeric:
        # add size to current path and parent pathes
        for i in range(1,len(path)+1):
            local_path = '/'.join(path[:i])
            memo[str(local_path)] += int(seg[0])
        
```

```python
total = 0 
for k,v in memo.items():
    if v <= 100000:
        total += v
total
```
