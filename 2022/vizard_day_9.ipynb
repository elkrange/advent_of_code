```python

def compute_tail(tail_pos):
    x,y = head_pos
    x2,y2 = tail_pos
    if abs(x-x2)>1 or abs(y-y2)>1:
        tail_pos = last_head
        pos_record.add(tail_pos)
    return tail_pos

for move in input_data.split('\n'):
    direction, count = move.split()
    for i in range(int(count)):
        last_head = head_pos
        x,y = head_pos
        match direction:
            case 'U':
                head_pos = (x,y+1)
            case 'D':
                head_pos = (x,y-1)
            case 'L':
                head_pos = (x-1,y)
            case 'R':
                head_pos = (x+1,y)
        tail_pos = compute_tail(tail_pos)

len(pos_record)
```
