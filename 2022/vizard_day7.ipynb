# Tree Build

```python
class Folder:
    def __init__(self, name,parent):
        self.name=name
        self.folders = []
        self.files_size=0
        self.parent=parent
    def add_file(self,size):
        self.files_size += size
    def add_folder(self,folder):
        self.folders.append(folder)
    def get_size(self):
        return self.files_size+ sum([folder.get_size() for folder in self.folders])
    def enter_dir(self,name):
        for folder in self.folders:
            if folder.name==name:
                return folder
    def get_tree(self):
        tree = [(self.name,self.get_size())]
        for folder in self.folders:
            tree.extend(folder.get_tree())
        return tree

root=Folder("/",None)
current_dir=root

for line in input_data.split("\n"):
    if line == "$ cd /":
        current_dir=root
        continue
    if line.startswith("$ cd") or line=='':
        if line=="$ cd ..":
            current_dir = current_dir.parent
            continue
        current_dir=current_dir.enter_dir(line[5:])
    elif line=="$ ls":
        continue
    else:
        if line[0].isnumeric():
            current_dir.add_file(int(line.split(' ')[0]))
        else:
            current_dir.add_folder(Folder(line.split(' ')[1],current_dir))



```
