```python
cycle = 0
x = 1
strength = 0
next_stop = 20
for i in input_data.split('\n'):
    if i == 'noop':
        cycle+=1
        if cycle == next_stop:
            strength+=x*next_stop
            next_stop += 40
            if next_stop==260:
                break
    else:
        _, val = i.split()
        val=int(val)
        if cycle+2 >= next_stop:
            strength+=x*next_stop
            next_stop += 40
            cycle+=2
            x+=val
            if next_stop==260:
                break
        
        else:
            x+=val
            cycle+=2
strength        
```
